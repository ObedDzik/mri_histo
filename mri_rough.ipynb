```python
DINOV3_LIBRARY_PATH = '/home/obed/projects/aip-medilab/obed/medproj/dinov3'
DINOV3_CHECKPOINTS_PATH='/datasets/exactvu_pca/checkpoint_store'
```

```python
def dinov3_vitl16(**kwargs):
    import torch
    import os
    REPO_DIR = DINOV3_LIBRARY_PATH
    weight_basename = "dinov3_vitl16_pretrain_lvd1689m-8aa4cbdd.pth"
    weights = os.path.join(DINOV3_CHECKPOINTS_PATH, weight_basename)
    if not os.path.exists(weights):
        raise FileNotFoundError(f"DINOv3 checkpoint not found: {weights}")
    kwargs["weights"] = weights
    model = torch.hub.load(REPO_DIR, "dinov3_vitl16", source="local", **kwargs)
    return model
```

```python
model = dinov3_vitl16()
```

```python
from create_model_wrapper import DINOModelWrapper
```

```python
import torch

dmodel = DINOModelWrapper(
    backbone=model,
    num_classes=6,
    img_size=256,
    proj_dim=512,
    attn_dim=512,
    head_hidden=512,
    head_dropout=0.1,
    pixel_mean_std=([0.485, 0.456, 0.406], [0.229, 0.224, 0.225]),
)


```

```python
for param in dmodel.parameters():
    param.requires_grad = False
for param in dmodel.encoder.parameters():
    param.requires_grad = True
for param in dmodel.pool.parameters():
    param.requires_grad = True
for param in dmodel.proj.parameters():
    param.requires_grad = True
```

```python

print(f"Number of parameters: {sum(p.numel() for p in dmodel.parameters())}")
print(f"Number of head parameters: {sum(p.numel() for p in dmodel.head.parameters())}")
print(f"Number of trainable parameters: {sum(p.numel() for p in dmodel.parameters() if p.requires_grad)}")
```

```python
optimizer = torch.optim.AdamW([
    {"params": dmodel.proj.parameters(), "lr": 1e-3},
    {"params": dmodel.pool.parameters(), "lr": 5e-3},   # 10x higher
    {"params": dmodel.encoder.parameters(), "lr": 1e-2}, # 5x higher
])
```

```python
from triplet_loss_utils import get_histo_by_isup, triplet_loss_batch
def train_triplet_loss(embeddings, labels):
    return triplet_loss_batch(
        embeddings = embeddings,
        labels = labels, 
        histo_dict = train_buckets, 
        margin = 1,
        reduction = 'mean',
        num_classes = 6, 
    )
```

```python
# Make sure ALL encoder params are trainable
for name, param in dmodel.encoder.named_parameters():
    if not param.requires_grad:
        print(f"FROZEN: {name}")
```

```python
train_loader[1]
```

```python

is_training = optimizer is not None
dmodel.train(is_training)
total_loss = 0.0
dmodel.to('cuda')
encoder_grad_norms = []
i=0
# Before training
with torch.no_grad():
    emb_before = dmodel(next(iter(train_loader))['image'].to('cuda'))

with torch.set_grad_enabled(is_training):
    for batch in tqdm(train_loader, desc='train'):
        i=i+1
        images = batch["image"].to('cuda', non_blocking=True)
        labels = batch["label"].to('cuda', non_blocking=True)
    
        _, embeddings = dmodel(images)
        loss = train_triplet_loss(embeddings, labels)
        if is_training:
            optimizer.zero_grad(set_to_none=True)
            loss.backward()

            for param in dmodel.encoder.parameters():
                if param.grad is not None:
                    encoder_grad_norms.append(param.grad.norm().item())

            if encoder_grad_norms:
                print(f"Encoder grad - mean: {np.mean(encoder_grad_norms):.4e}, "
                    f"max: {np.max(encoder_grad_norms):.4e}, "
                    f"min: {np.min(encoder_grad_norms):.4e}")
            else:
                print("NO GRADIENTS in encoder")

            proj_grad_norms = [p.grad.norm().item() for p in dmodel.proj.parameters() if p.grad is not None]
            pool_grad_norms = [p.grad.norm().item() for p in dmodel.pool.parameters() if p.grad is not None]

            print(f"Proj grad - mean: {np.mean(proj_grad_norms):.4e}, max: {np.max(proj_grad_norms):.4e}, min: {np.min(proj_grad_norms):.4e}")
            print(f"Pool grad - mean: {np.mean(pool_grad_norms):.4e}, max: {np.max(pool_grad_norms):.4e}, min: {np.max(pool_grad_norms):.4e}")

            torch.nn.utils.clip_grad_norm_(dmodel.parameters(), max_norm=1.0)
            optimizer.step()
        batch_size = images.size(0)
        total_loss += loss.item()
        print("Total Loss: ", loss.item())

        if i == 10:  # only 5 batches
            break
```
